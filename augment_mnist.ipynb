```python
from keras.datasets import mnist
import numpy as np
import cv2


(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
SIZE_TRAIN = x_train.shape[0]
SIZE_TEST = x_test.shape[0]
threshed_trains = np.zeros(x_train.shape)
for i in range(SIZE_TRAIN):
    _, threshed_trains[i] = cv2.threshold(x_train[i], 100, 255, cv2.THRESH_BINARY)

threshed_tests =  np.zeros(x_test.shape)
for i in range(SIZE_TEST):
    _, threshed_tests[i] = cv2.threshold(x_test[i], 100, 255, cv2.THRESH_BINARY)

resized_trains = np.zeros(x_train.shape)
for i in range(SIZE_TRAIN):
    new_img = np.zeros((28, 28))
    if bool(np.random.randint(0, 2)):
        x, y = np.random.randint(0, 2, 2)
        img = cv2.resize(x_train[i], (26, 26))
        new_img[x:(-2+x),y:(-2+y)] = img
    else:
        x, y = np.random.randint(0, 4, 2)
        img = cv2.resize(x_train[i], (24, 24))
        new_img[x:(-4+x),y:(-4+y)] = img
    resized_trains[i] = new_img

resized_tests = np.zeros(x_test.shape)
for i in range(SIZE_TEST):
    new_img = np.zeros((28, 28))
    if bool(np.random.randint(0, 2)):
        x, y = np.random.randint(0, 2, 2)
        img = cv2.resize(x_test[i], (26, 26))
        new_img[x:(-2+x),y:(-2+y)] = img
    else:
        x, y = np.random.randint(0, 4, 2)
        img = cv2.resize(x_test[i], (24, 24))
        new_img[x:(-4+x),y:(-4+y)] = img
    resized_tests[i] = new_img

x_train_a = np.zeros((180000, 28, 28))
x_train_a[:SIZE_TRAIN] = x_train
x_train_a[SIZE_TRAIN:SIZE_TRAIN*2] = threshed_trains
x_train_a[SIZE_TRAIN*2:] = resized_trains
x_test_a = np.zeros((30000, 28, 28), dtype='int4')
x_test_a[:SIZE_TEST] = x_test
x_test_a[SIZE_TEST:SIZE_TEST*2] = threshed_tests
x_test_a[SIZE_TEST*2:] = resized_tests

y_train_a = np.zeros((SIZE_TRAIN*3))
y_train_a[:SIZE_TRAIN] = y_train
y_train_a[SIZE_TRAIN:SIZE_TRAIN*2] = y_train
y_train_a[SIZE_TRAIN*2:] = y_train
y_test_a = np.zeros((SIZE_TEST*3), dtype='int4')
y_test_a[:SIZE_TEST] = y_test
y_test_a[SIZE_TEST:SIZE_TEST*2] = y_test
y_test_a[SIZE_TEST*2:] = y_test
x_train_a = x_train_a.reshape((SIZE_TRAIN*3,28*28))
x_test_a = x_test_a.reshape((SIZE_TEST*3,28*28))
idx = np.argsort(np.random.random(SIZE_TRAIN))
x_train_a = x_train_a[idx]
y_train_a = y_train_a[idx]

idx = np.argsort(np.random.random(SIZE_TEST))
x_test_a = x_test_a[idx]
y_test_a = y_test_a[idx]

np.save('./data/mnist/super_augmented/train_vectors.npy', x_train_a.astype(int))
np.save('./data/mnist/super_augmented/train_labels.npy', y_train_a.astype(int))
np.save('./data/mnist/super_augmented/test_vectors.npy', x_test_a.astype(int))
np.save('./data/mnist/super_augmented/test_labels.npy', y_test_a.astype(int))
```
